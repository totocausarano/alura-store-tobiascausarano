### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt


url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head(5)
```

#1. Análisis de facturación



```python
ingreso_total1 = tienda["Precio"].sum()
ingreso_total2 = tienda2['Precio'].sum()
ingreso_total3 = tienda3['Precio'].sum()
ingreso_total4 = tienda4['Precio'].sum()
```

```python
print(ingreso_total1)
print(ingreso_total2)
print(ingreso_total3)
print(ingreso_total4)
```

```python
valores = [ingreso_total1, ingreso_total2, ingreso_total3, ingreso_total4]
etiquetas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
colores = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']
plt.pie(valores, labels=etiquetas, startangle = 40, colors=colores, autopct= '%1.1f%%')

plt.title('Ingresos por tienda')

plt.show()
```

# 2. Ventas por categoría

```python
categoria1 = tienda.groupby('Categoría del Producto')['Precio'].count()
categoria2 = tienda2.groupby('Categoría del Producto')['Precio'].count()
categoria3 = tienda3.groupby('Categoría del Producto')['Precio'].count()
categoria4 = tienda4.groupby('Categoría del Producto')['Precio'].count()
```

```python
display(categoria1)
display(categoria2)
display(categoria3)
display(categoria4)
```

```python
df_cantidades = pd.DataFrame({'Tienda 1' : categoria1, 'Tienda 2' : categoria2, 'Tienda 3' : categoria3, 'Tienda 4' : categoria4})
df_cantidades.plot(kind='bar', figsize=(12, 6), width = 0.8)
plt.title('Cantidad de productos por categoría')
plt.xlabel('Categoría')
plt.ylabel('Cantidad')
plt.show()
```

# 3. Calificación promedio de la tienda


```python
calificacion_media1 = tienda['Calificación'].sum()/len(tienda)
calificacion_media2 = tienda2['Calificación'].sum()/len(tienda2)
calificacion_media3 = tienda3['Calificación'].sum()/len(tienda3)
calificacion_media4 = tienda4['Calificación'].sum()/len(tienda4)
```

```python
display(calificacion_media1)
display(calificacion_media2)
display(calificacion_media3)
display(calificacion_media4)
```

```python
df_calificaciones = pd.Series({'Tienda 1' : calificacion_media1, 'Tienda 2' : calificacion_media2, 'Tienda 3' : calificacion_media3, 'Tienda 4' : calificacion_media4})
df_calificaciones.plot(kind='barh', figsize=(12, 6), width = 0.3)
plt.title('Calificación promedio por tienda')
plt.xlabel('Calificación')
plt.xlim(0,5)
plt.show()
```

# 4. Productos más y menos vendidos

```python
productos1 = tienda['Producto'].value_counts()
productos2 = tienda2['Producto'].value_counts()
productos3 = tienda3['Producto'].value_counts()
productos4 = tienda4['Producto'].value_counts()
```

```python
display(productos1)
display(productos2)
display(productos3)
display(productos4)
```

#Informe final:

El objetivo de este trabajo es analizar los datos de las cuatro tiendas de la cadena para decidir cuál nos conviene vender. Para esto, comparé cómo le fue a cada una revisando cuánto facturaron, qué productos vendieron más, qué calificaciones les pusieron los clientes y cuánto gastaron en envíos. La idea es usar estos números para justificar, con pruebas concretas, cuál es la tienda que menos aporta al negocio hoy en día.

##Desarrollo del Analisis:

 **Facturación e Ingresos Totales**

Al analizar los ingresos brutos generados por cada tienda, se observa una diferencia significativa entre la tienda con mejor rendimiento y la de menor rendimiento:

Tienda 1: $1,150,880,400 (Líder en ingresos)

Tienda 2: $1,116,343,500

Tienda 3: $1,098,019,600

Tienda 4: $1,038,375,700 (Menor ingreso)

Observación: La Tienda 1 es el motor financiero del grupo, superando a la Tienda 4 por más de $112 millones. La Tienda 4 representa la menor cuota de mercado en términos monetarios.

**Ventas por Categoría y Volumen**

Aunque la facturación varía, el comportamiento de las categorías es consistente. Las categorías dominantes en todas las tiendas son Muebles y Electrónicos.

Sin embargo, al observar los productos más vendidos, notamos una diferencia en el "ticket promedio".

La Tienda 1 tiene como top ventas productos de alto valor como "Microondas" y "TV LED UHD 4K".

La Tienda 4, aunque mueve volumen, tiene entre sus productos más vendidos la "Cubertería", un artículo de menor valor unitario, lo que explica por qué su facturación es menor a pesar de tener un número de transacciones decente.

**Satisfacción del Cliente (Calificación)**

La calidad del servicio es::

Tienda 3: 4.05 estrellas (Mejor valorada).

Tienda 2: 4.04 estrellas.

Tienda 4: 3.99 estrellas.

Tienda 1: 3.97 estrellas.

Aunque la Tienda 1 es la que más factura, es la que tiene la calificación más baja, lo cual es un punto de atención operativa, pero no financiero. La Tienda 4 se mantiene en un promedio aceptable, sin destacar.

**Costos Logísticos (Envío Promedio)**

Se analizó la eficiencia en los costos de envío promedio por producto:

Tienda 1: ~$26,018 (El más alto, probablemente por vender productos más voluminosos).

Tienda 4: ~$23,459 (El más bajo).

La Tienda 4 es la más eficiente en envíos, pero esto probablemente se debe a que vende productos más pequeños o de menor valor (como se vio en el análisis de productos), y no necesariamente a una mejor gestión logística.

##**Conclusión y Recomendación**
Basado en el análisis integral de los datos, la recomendación estratégica es vender la Tienda 4.

Justificación de la decisión:
Menor Rendimiento Financiero: La Tienda 4 es, consistentemente, la que genera menos ingresos ($1.03 mil millones), quedándose significativamente atrás de la Tienda 1 y la Tienda 2. En una estrategia de consolidación, es la que menos aporta.

A diferencia de las otras tiendas que lideran con tecnología y electrodomésticos, el volumen de la Tienda 4 se apoya más en artículos de hogar de menor costo (como cubertería y camas box), lo que exige más esfuerzo logístico para igualar la facturación de las demas tiendas

Posición de Mercado: No es líder en satisfacción al cliente (superada por Tienda 2 y 3) ni líder en ventas (superada por Tienda 1). Se encuentra en una posición estancada.

Acción Sugerida: Proceder con la venta de la Tienda 4 y reinvertir el capital y los esfuerzos operativos en mejorar la experiencia de usuario de la Tienda 1 (que tiene la calificación más baja) para potenciar aún más.
